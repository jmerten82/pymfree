```python
def convert_terms(terms,dims):
    out = torch.zeros([len(terms),len(dims)],dtype=torch.int64)
    for i,term in enumerate(terms):
        for j, dim in enumerate(dims):
            out[i,j] = (term==dim).sum()
    return out
```

```python
def convert_terms_faster(terms,dims):
    stack=[]
    for dim in dims:
        stack.append((terms==dim).sum(axis=1))
    return torch.stack(stack,1)
```
